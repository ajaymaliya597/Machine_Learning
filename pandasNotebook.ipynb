```python
# ==========================================================
# GROUPBY
# ==========================================================

# Count Passengers by Gender
df.groupby("Sex").size()

# Count Passengers by Passenger Class
df.groupby("Pclass").size()

# Average Age by Gender
df.groupby("Sex")["Age"].mean()

# Average Fare by Passenger Class
df.groupby("Pclass")["Fare"].mean()

# Total Fare by Passenger Class
df.groupby("Pclass")["Fare"].sum()

# Maximum Fare by Passenger Class
df.groupby("Pclass")["Fare"].max()

# Minimum Fare by Passenger Class
df.groupby("Pclass")["Fare"].min()

# Survival Count by Gender
df.groupby("Sex")["Survived"].sum()

# Average Age by Passenger Class
df.groupby("Pclass")["Age"].mean()

```

```python

# ==========================================================
# CONCATENATION
# ==========================================================

# Vertical Concatenation

df1 = pd.DataFrame({
    "Emp_ID":[101,102],
    "Name":["Vinay","Aman"]
})

df2 = pd.DataFrame({
    "Emp_ID":[103,104],
    "Name":["Riya","Priya"]
})

vertical = pd.concat([df1,df2])

print(vertical)

```

```python

# ==========================================================
# Horizontal Concatenation
# ==========================================================

employee = pd.DataFrame({
    "Emp_ID":[101,102,103],
    "Name":["Vinay","Aman","Riya"]
})

salary = pd.DataFrame({
    "Salary":[50000,45000,60000],
    "Bonus":[5000,3000,7000]
})

horizontal = pd.concat(
    [employee,salary],
    axis=1
)

print(horizontal)
```

```python

# ==========================================================
# MERGE
# ==========================================================

employees = pd.DataFrame({
    "Emp_ID":[101,102,103,104],
    "Name":["Vinay","Aman","Riya","Priya"]
})

salary = pd.DataFrame({
    "Emp_ID":[101,102,105],
    "Salary":[50000,45000,60000]
})

# ==========================================================
# INNER JOIN
# ==========================================================

inner = pd.merge(
    employees,
    salary,
    on="Emp_ID",
    how="right"
)

print(inner)
```

```python

# ==========================================================
# LEFT JOIN
# ==========================================================

left = pd.merge(
    employees,
    salary,
    on="Emp_ID",
    how="left"
)

print(left)

# ==========================================================
# RIGHT JOIN
# ==========================================================

right = pd.merge(
    employees,
    salary,
    on="Emp_ID",
    how="right"
)

print(right)

# ==========================================================
# OUTER JOIN
# ==========================================================

outer = pd.merge(
    employees,
    salary,
    on="Emp_ID",
    how="outer"
)

print(outer)
```

```python

# Create Sample Date Column

df["Booking_Date"] = [
    "2024-01-01","2024-01-05","2024-01-10","2024-01-12",
    "2024-01-15","2024-01-18","2024-01-20","2024-01-25",
    "2024-01-28","2024-02-01"
] * (len(df)//10) + ["2024-02-05"]*(len(df)%10)

# Convert into Datetime

df["Booking_Date"] = pd.to_datetime(df["Booking_Date"])

# Extract Year
df["Year"] = df["Booking_Date"].dt.year

# Extract Month
df["Month"] = df["Booking_Date"].dt.month

# Extract Day
df["Day"] = df["Booking_Date"].dt.day

# Extract Day Name
df["Day_Name"] = df["Booking_Date"].dt.day_name()

# Extract Month Name
df["Month_Name"] = df["Booking_Date"].dt.month_name()

print(df[["Booking_Date","Year","Month","Day","Day_Name"]].head())
```

```python

# ==========================================================
# APPLY()
# ==========================================================

# Create Age Category

def age_category(age):
    if pd.isnull(age):
        return "Unknown"
    elif age < 18:
        return "Child"
    elif age < 60:
        return "Adult"
    else:
        return "Senior Citizen"

df["Age_Category"] = df["Age"].apply(age_category)
```

```python
df['Double_age']=df["Age"].apply(lambda x:2*x)
```

```python

# ==========================================================
# MAP()
# ==========================================================

# Convert Gender into Numbers

gender_map = {
    "male":0,
    "female":1
}

df["Gender"] = df["Sex"].map(gender_map)

# Passenger Class Mapping

class_map = {
    1:"First Class",
    2:"Second Class",
    3:"Third Class"
}

df["Passenger_Class"] = df["Pclass"].map(class_map)

print(df[["Sex","Gender","Pclass","Passenger_Class"]].head())

```

```python
# ==========================================================
# APPLYMAP()
# ==========================================================

# Create Sample DataFrame

marks = pd.DataFrame({
    "Math":[80,70,90],
    "Science":[85,75,95],
    "English":[88,78,92]
})

print(marks)

# Increase Every Value by 5

marks = marks.applymap(lambda x: x+5)

print(marks)

# Convert Every Value into Percentage

marks = marks.applymap(lambda x: str(x)+" %")

print(marks)
```

```python

# ==========================================================
# LAMBDA FUNCTIONS
# ==========================================================

# Passenger Type

df["Passenger_Type"] = df["Age"].apply(
    lambda x: "Child" if x < 18 else "Adult"
)

# High Fare Passenger

df["Premium"] = df["Fare"].apply(
    lambda x: "Yes" if x > 100 else "No"
)

# Family or Solo Traveller

df["Traveller"] = (df["SibSp"] + df["Parch"]).apply(
    lambda x: "Family" if x > 0 else "Solo"
)

print(df[["Age","Passenger_Type","Fare","Premium","Traveller"]].head())

```
